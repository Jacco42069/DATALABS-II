```python
def Stats(df, t_id, season):
    #selecteer de club
    wedstrijd = df[((df['home_team_api_id'] == t_id) | (df['away_team_api_id'] == t_id)) & (df['season'] == season)]
    
    #bepaal het aantal wedstrijden dat de club heeft gespeeld
    
    
    #bereken het aantal punten behaald en het gemiddelde aantal punten per seizoen.
    aantal_wedstrijden = len(wedstrijd)
    punten = 0
    for index, wedstrijden in wedstrijd.iterrows():
        if wedstrijden['home_team_api_id'] == t_id:
            if wedstrijden['home_team_goal'] > wedstrijden['away_team_goal']:
                punten += 3
            elif wedstrijden['home_team_goal'] == wedstrijden['away_team_goal']:
                punten += 1
        else:
            if wedstrijden['away_team_goal'] > wedstrijden['home_team_goal']:
                punten += 3
            elif wedstrijden['away_team_goal'] == wedstrijden['home_team_goal']:
                punten += 1
    gemiddeld_punten_per_wedstrijd = punten / aantal_wedstrijden
    
    thuis = wedstrijd[wedstrijd['home_team_api_id'] == t_id]
    uit = wedstrijd[wedstrijd['away_team_api_id'] == t_id]

    
    aantal_thuiswedstrijden = len(thuis)
    aantal_uitwedstrijden = len(uit)

    gewonnen_thuis = len(thuis[thuis['home_team_goal'] > thuis['away_team_goal']])
    verloren_thuis = len(thuis[thuis['home_team_goal'] < thuis['away_team_goal']])
    gelijk_thuis = len(thuis[thuis['home_team_goal'] == thuis['away_team_goal']])

    gewonnen_uit = len(uit[uit['away_team_goal'] > uit['home_team_goal']])
    verloren_uit = len(uit[uit['away_team_goal'] < uit['home_team_goal']])
    gelijk_uit = len(uit[uit['away_team_goal'] == uit['home_team_goal']])

    # Bereken het winst-, verlies- en gelijk percentage voor thuis- en uitwedstrijden
    winst_perc_thuis = gewonnen_thuis / aantal_thuiswedstrijden * 100
    verlies_perc_thuis = verloren_thuis / aantal_thuiswedstrijden * 100
    gelijk_perc_thuis = gelijk_thuis / aantal_thuiswedstrijden * 100

    winst_perc_uit = gewonnen_uit / aantal_uitwedstrijden * 100
    verlies_perc_uit = verloren_uit / aantal_uitwedstrijden * 100
    gelijk_perc_uit = gelijk_uit / aantal_uitwedstrijden * 100
    
    resultaten = {'team id': t_id,
                  'gemiddeld punten': gemiddeld_punten_per_wedstrijd, 
                  'totaal punten':punten, 
                  'winst thuis in %': winst_perc_thuis,
                  'winst uit in %': winst_perc_uit,
                  'verlies thuis in %': verlies_perc_thuis, 
                  'verlies uit in %': verlies_perc_uit, 
                  'gelijk thuis in %': gelijk_perc_thuis, 
                  'gelijk uit in %': gelijk_perc_uit}
    return resultaten, wedstrijd
```

```python
def Ranglijst(df, season):
    # Maak een lege dictionary aan om de punten per team bij te houden
    punten_per_team = {}

    # Loop over alle wedstrijden van het seizoen
    for index, wedstrijd in df[df['season'] == season].iterrows():
        home_team = wedstrijd['home_team_api_id']
        away_team = wedstrijd['away_team_api_id']
        resultaten_home, _ = Stats(df, home_team, season)
        resultaten_away, _ = Stats(df, away_team, season)

        # Voeg de behaalde punten van de thuis- en uitploeg toe aan de dictionary
        if home_team in punten_per_team:
            punten_per_team[home_team] += resultaten_home['totaal punten']
        else:
            punten_per_team[home_team] = resultaten_home['totaal punten']

        if away_team in punten_per_team:
            punten_per_team[away_team] += resultaten_away['totaal punten']
        else:
            punten_per_team[away_team] = resultaten_away['totaal punten']

    # Sorteer de dictionary op basis van de behaalde punten
    sorted_punten = sorted(punten_per_team.items(), key=lambda x: x[1], reverse=True)

    # Print de ranglijst
    print("Ranglijst seizoen " + season + ":")
    for i, (team_id, punten) in enumerate(sorted_punten):
        team_naam = df[df['home_team_api_id'] == team_id]['home_team_api_id'].values[0]
        print(str(i+1) + ". " + team_naam + " (" + str(punten) + " punten)")
```
